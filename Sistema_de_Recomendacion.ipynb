# SISTEMA DE RECOMENDACIÓN DE VIDEOJUEGOS BASADOS EN LA PLATAFORMA STEAM

```python
# IMPORTACIÓN DE LIBRERIAS...

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import os
```

```python
# LECTURA DEL DATAFRAME DE TRABAJO...
SystemTheRecommendation = pd.read_csv("Archivos_para_API/SystemTheRecommendation.csv")
```

```python
# ELIMINACIÓN DE DUPLICADOS...
SystemTheRecommendation = SystemTheRecommendation.drop_duplicates()
```

```python
# Crear la matriz pivotante
SystemTheRecommendation_matriz = SystemTheRecommendation.pivot_table(index='item_name', columns='user_id', values='rating', fill_value=0)
SystemTheRecommendation_matriz = SystemTheRecommendation_matriz.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)
SystemTheRecommendation_matriz = SystemTheRecommendation_matriz.dropna(how='all')

# Agregar la columna 'item_name' al DataFrame resultante
SystemTheRecommendation_matriz['item_name'] = SystemTheRecommendation_matriz.index

# Restablecer el índice para que 'item_name' sea una columna
SystemTheRecommendation_matriz.reset_index(drop=True, inplace=True)
```

```python
# TRABAJO PARA LA RECOMENDACIÓN SEGÚN EL USUARIO...
SystemTheRecommendation_matriz_user = SystemTheRecommendation.pivot_table(index='item_name', columns='user_id', values='rating', fill_value=0)
SystemTheRecommendation_matriz_user = SystemTheRecommendation_matriz_user.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)
SystemTheRecommendation_matriz_user = SystemTheRecommendation_matriz_user.dropna(how='all')
```

```python
# TRABAJO PARA LA RECOMENDACIÓN SEGÚN EL NOMBRE DEL VIDEOJUEGO...
SystemTheRecommendation_matriz_item = SystemTheRecommendation.pivot_table(index='user_id', columns='item_name', values='rating', fill_value=0)
SystemTheRecommendation_matriz_item = SystemTheRecommendation_matriz_item.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)
SystemTheRecommendation_matriz_item = SystemTheRecommendation_matriz_item.dropna(how='all')
```

```python
# Aplicar la similitud del coseno
cosine_sim = cosine_similarity(SystemTheRecommendation_matriz_user.T)  # .T transpone el DataFrame para tener usuarios en filas

# Crear un nuevo DataFrame con los resultados de la similitud del coseno
cosine_sim_user = pd.DataFrame(cosine_sim, index=SystemTheRecommendation_matriz_user.columns, columns=SystemTheRecommendation_matriz_user.columns)
```

```python
# Aplicar la similitud del coseno
cosine_sim = cosine_similarity(SystemTheRecommendation_matriz_item.T)  # .T transpone el DataFrame para tener usuarios en filas

# Crear un nuevo DataFrame con los resultados de la similitud del coseno
cosine_sim_item = pd.DataFrame(cosine_sim, index=SystemTheRecommendation_matriz_item.columns, columns=SystemTheRecommendation_matriz_item.columns)
```
